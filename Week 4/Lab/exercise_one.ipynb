```python
import pandas as pd
import scipy as sp

clean_crime_csv= pd.read_csv('censusCrimeClean.csv')

df = pd.DataFrame(clean_crime_csv)
```

```python
df.dtypes
```

```python
median_income = df.medIncome
violent_crimes_pp = df.ViolentCrimesPerPop
household_size = df.householdsize
pct_black = df.racepctblack
pct_white = df.racePctWhite
pct_asian = df.racePctAsian
pct_hisp = df.racePctHisp
```

```python
print(sp.stats.pearsonr(median_income, violent_crimes_pp))

print(sp.stats.pearsonr(pct_black, violent_crimes_pp))
print(sp.stats.pearsonr(pct_white, violent_crimes_pp))
print(sp.stats.pearsonr(pct_asian, violent_crimes_pp))
print(sp.stats.pearsonr(pct_hisp, violent_crimes_pp))
```

```python
print(sp.stats.pearsonr(median_income, violent_crimes_pp))
sp.stats.spearmanr(median_income, violent_crimes_pp)
# Results show that violent crime goes down as median income goes up
```
