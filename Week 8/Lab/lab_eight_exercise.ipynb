```python
from urllib.request import urlopen
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Princess Badoura, A tale from the Arabian Nights. By Laurence Housman
book_url = "https://www.gutenberg.org/files/51219/51219-0.txt"

book_raw = urlopen(book_url).read().decode('utf-8')
```

```python
print(f'Length of book_raw: {len(book_raw)}')
```

```python
word_tokens = word_tokenize(book_raw)
# print(word_tokens[1:40])
```

```python
stop_words = set(stopwords.words('english'))
# print(stop_words)
```

```python
smaller_token_list = word_tokens[10000:20000]
# print(smaller_token_list)
```

```python
book_sentences = sent_tokenize(book_raw)
print(f'Princess Badoura has {len(book_sentences)} sentences throughout the book.')
```

```python
sentence_length_list = []
for s in book_sentences:
    sentence_length_list.append(len(s))    

print(sentence_length_list)
```

```python
number_of_sentences = len(sentence_length_list)

sum_sentence_chars = sum(sentence_length_list)

average_char_sentences = round(sum_sentence_chars/number_of_sentences, 0)

print(f'The average number of characters in a sentence is: {average_char_sentences}')
```

```python
from nltk import FreqDist

fdist_example = FreqDist(word_tokens)
# fdist_example

print(f'The number of unique words in the book is: {len(fdist_example)}')
```

```python
import matplotlib


wordcloud = WordCloud(width=1000, height=700, background_color='black',colormap=matplotlib.cm.autumn,
                min_font_size = 10).generate(str(word_tokens))

matplotlib.pyplot.imshow(wordcloud)
```

```python
uppercase_list = []
for word in word_tokens:
    if word[0].isupper() == True:
        uppercase_list.append(word)
```

```python
#print(uppercase_list)
print(len(uppercase_list))
```

```python
uppercase_dist = FreqDist(uppercase_list)

uppercase_dist
```

```python
uppercase_wordcloud = WordCloud(width=1000, height=700, background_color='black',colormap=matplotlib.cm.autumn,
                min_font_size = 10).generate(str(uppercase_list))

matplotlib.pyplot.imshow(uppercase_wordcloud)
```
